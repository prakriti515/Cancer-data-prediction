```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
cancer_data=pd.read_csv("Cancer.csv")
cancer_data.head()
```

```python
cancer_data.shape
```

```python
cancer_data.describe()
```

```python
null_values = cancer_data.isnull().sum()
null_values
```

```python
cancer_data.info()
```

```python
cancer_data.columns
```

```python
# counts of unique rows in the 'diagnosis' column
cancer_data['diagnosis'].value_counts()
```

```python
# dropping 'Unnamed: 32' column.
cancer_data.drop("Unnamed: 32", axis=1, inplace=True)
```

```python
# dropping id column
cancer_data.drop('id', axis=1, inplace=True)
```

```python
Y = cancer_data['diagnosis']
X = cancer_data.drop('diagnosis', axis=1)
```

```python
cancer_data.select_dtypes(include='number').corr()
```

```python
import seaborn as sns
correlation_matrix = cancer_data.select_dtypes(include='number').corr()

plt.figure(figsize=(20,18))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
# Splitting the dataset into 80/20 ratio 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
print(Y.shape, Y_test.shape, Y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
model.fit(X_train,Y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
Y_predict = model.predict(X_test)
Y_predict
```

```python
confusion_matrix(Y_test,Y_predict)
```

```python
accuracy_score(Y_test,Y_predict)
```

```python
print(classification_report(Y_test,Y_predict))
```

```python
log_acc1 = accuracy_score(Y_test, Y_predict)
print("Accuracy of linear regression model is: ", log_acc1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.8, random_state=2529)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
Y_predict = model.predict(X_test)
Y_predict
```

```python
accuracy_score(Y_test,Y_predict)
print(classification_report(Y_test,Y_predict))
```

```python
log_acc2 = accuracy_score(Y_test, Y_predict)
print("Accuracy of linear regression model is: ", log_acc2)
```

```python
from sklearn.svm import SVC

svc_model = SVC(kernel="rbf")
svc_model.fit(X_train, Y_train)
predictions = svc_model.predict(X_test)
```

```python
print("Confusion Matrix: \n", confusion_matrix(Y_test, predictions))
print("\n")
print(classification_report(Y_test, predictions))
```

```python
svm_acc = accuracy_score(Y_test, predictions)
print("Accuracy of SVM model is: ", svm_acc)
```

```python
print(log_acc1)
print(log_acc2)
print(svm_acc)
```

```python
plt.figure(figsize=(12,6))
model_acc = [log_acc1, log_acc2, svm_acc]
model_name = ['LogisticRegression on 80,20', 'LogisticRegression on 20,80', 'SVM']
sns.barplot(x= model_acc, y=model_name)
```
